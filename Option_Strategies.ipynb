```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
S = np.linspace(6750,8350,33 )  # This Creates the Spot prices as a range
```

```python
S
```

```python
# Call European Option
# Here, as we know, if in call option price goes up, we make profit as future price less the strike price less the Premium , and if call option goes down we make a loss equal to premium. Premium = call value when we buy the option contract

def CE(St,k,v):
  payoff = []
  for i in St:
    p = np.maximum(i-k,0)
    payoff.append(p-v)
  return (np.array(payoff))
```

```python
CE(S,7600,88)  # This function returns the net Payoff for Call Option
```

```python
# Put European Option
# Here, as we know, if in put option price goes down, we make profit as strike price less the future price less the Premium , and if call option goes up we make a loss equal to premium. Premium = put value when we buy the option contract

def PE(St,k,v):
  payoff = []
  for i in St:
    p = np.maximum(k-i,0)
    payoff.append(p-v)
  return (np.array(payoff))
```

```python
PE(S,7600,77) # This Function returns the net Payoff for Put Option
```

```python
zero = np.zeros(S.shape)  # this is creteed to highlight y = 0 line

def Plot(arr_x,arr_y,x_name,y_name,title):   # Very Basic structure followed to Plot the Payoff diagram using plt.plot()
  plt.figure(figsize = (12,9))
  plt.plot(arr_x,arr_y)
  plt.plot(arr_x,zero,color="black")
  plt.xlabel(x_name)
  plt.ylabel(y_name)
  plt.title(title)
  plt.grid()
  plt.show()
```

```python
Plot(S, CE(S,7600,88), "stock price","Payoff","Call Option Payoff")
```

```python
Plot(S, PE(S,7600,77), "stock price","Payoff","Put Option Payoff")
```

```python
# LONG STRADDLE STRATEGY


# In this strategy, we buy both call and put option having the same underlying Asset
# Having the same underlying Asset would mean the exact same strike price and Spot prices
# This strategy is developed to counter the market's movement direction. That means irrespesctive of whether the market moves up or down , it must provide profits.
# Assumption used : Market always moves and doesn't stay at the same level , so in either direction profit made would offset the net Premium.

# Buying both call and Put option ;So, adding up the payoffs to get the net Profit


call =CE(S,7600,88)
put = PE(S,7600,77)
call + put
```

```python
Plot(S,call + put,"Stock price","Payoff", "Net Payoff Diagram")
```
